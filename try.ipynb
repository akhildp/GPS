### Import packages

```python
import urllib2
import multiprocessing as mp
```

### We create a test list of 1000 url strings, that return a random quote when we make a get request. Then we populate every %100 element in the list with a bad url 'http://irgknjvdf' to check our exception handling

```python
url_list='http://ron-swanson-quotes.herokuapp.com/v2/quotes;'*1000
url_list=url_list.split(';')
for i in range(len(url_list)):
    if i%100==0:
        url_list[i]='http://irgknjvdf'
```

### As you can see, the very first element in the list is a bad url

```python
url_list
```

### The function that makes a get request has a try catch block to catch the URL errors we would encounter.

```python
def get_req(url):
    try:
        response = urllib2.urlopen(url)
        data = response.read()
        response.close() 
        return data
    except urllib2.URLError:
        return 'URL Error'
```

### Create a pool of 500 threads and make parallel requests to the urls.

```python
pool = mp.Pool(500)
results = pool.map(get_req,url_list)
```

### As you can see we hit the URL Error at every 100th url in the list, while we get valid response for all others

```python
results
```
